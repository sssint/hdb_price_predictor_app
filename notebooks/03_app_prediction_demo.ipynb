# HDB App Input and Prediction Demo

This notebook shows how the Streamlit app builds one prediction row with the same features used during model training.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent
sys.path.append(str(PROJECT_ROOT))

import pandas as pd
from catboost import CatBoostRegressor
from data_loader import clean_hdb_data
from train_model import FEATURES, MODEL_PATH
```

```python
# Example user input. The app.py UI collects these values from Streamlit widgets.
sample_raw = pd.DataFrame([{
    "month": "2025-06",
    "town": "TAMPINES",
    "flat_type": "4 ROOM",
    "storey_range": "07 TO 09",
    "floor_area_sqm": 93,
    "flat_model": "MODEL A",
    "lease_commence_date": 1990,
    "remaining_lease": "64 years 00 months",
    "resale_price": 500000,  # dummy value only so clean_hdb_data can run
}])

sample = clean_hdb_data(sample_raw)[FEATURES]
sample
```

```python
import numpy as np

if MODEL_PATH.exists():
    model = CatBoostRegressor()
    model.load_model(str(MODEL_PATH))
    prediction = model.predict(sample)[0]
    print("Predicted resale price SGD:", round(float(np.expm1(prediction)), 2))
else:
    print("Model file not found. Run train_model.py or the training notebook first.")
```
