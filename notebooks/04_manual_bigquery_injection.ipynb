## Step 3: Upload CSV to BigQuery
Change `YOUR_PROJECT_ID` to your Google Cloud project ID.

```python
!python bigquery/manual_upload_csv_to_bigquery.py \
  --project-id YOUR_PROJECT_ID \
  --dataset hdb_ml \
  --table resale_transactions_raw \
  --csv data/manual_upload/hdb_resale_from_data_gov.csv \
  --location asia-southeast1
```

## Step 4: Read from BigQuery into pandas

```python
from bigquery.bigquery_reader import read_table

PROJECT_ID = "YOUR_PROJECT_ID"
df = read_table(PROJECT_ID, dataset="hdb_ml", table="resale_transactions_raw")
df.head()
```
