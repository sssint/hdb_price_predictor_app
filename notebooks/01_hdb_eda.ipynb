# HDB Resale Price Prediction — EDA Notebook

This notebook explores the official HDB resale-flat dataset from data.gov.sg and checks the engineered features used by the Streamlit/Cloud Run app.

```python
# Run from the project root, not inside notebooks/
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent
sys.path.append(str(PROJECT_ROOT))

import pandas as pd
import matplotlib.pyplot as plt
from data_loader import fetch_hdb_resale_data, clean_hdb_data
```

```python
# Use max_records for quick testing. Set to None for full dataset.
raw = fetch_hdb_resale_data(max_records=5000)
raw.head()
```

```python
raw.shape, raw.columns.tolist()
```

```python
df = clean_hdb_data(raw)
df.head()
```

```python
df[["resale_price", "floor_area_sqm", "remaining_lease_years", "flat_age", "distance_to_nearest_mrt_km", "nearby_schools_2km"]].describe()
```

```python
plt.figure(figsize=(8, 5))
df["resale_price"].hist(bins=40)
plt.title("HDB Resale Price Distribution")
plt.xlabel("Resale Price SGD")
plt.ylabel("Count")
plt.show()
```

```python
town_avg = df.groupby("town")["resale_price"].mean().sort_values(ascending=False).head(15)
plt.figure(figsize=(10, 5))
town_avg.plot(kind="bar")
plt.title("Average Resale Price by Town - Top 15")
plt.ylabel("Average Price SGD")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(df["floor_area_sqm"], df["resale_price"], alpha=0.25)
plt.title("Floor Area vs Resale Price")
plt.xlabel("Floor Area sqm")
plt.ylabel("Resale Price SGD")
plt.show()
```
